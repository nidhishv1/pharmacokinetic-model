```python
import numpy as np
import matplotlib.pyplot as plt

# Time range (hours)
t = np.linspace(0, 24, 100)

# Model parameters
dose = 100            # mg
elimination_rate = 0.15  # 1/hour
volume = 10           # liters

# First-order elimination model
concentration = (dose / volume) * np.exp(-elimination_rate * t)

# Plot results
plt.plot(t, concentration)
plt.xlabel("Time (hours)")
plt.ylabel("Drug Concentration (mg/L)")
plt.title("Drug Concentration vs Time")
plt.show()

toxic_level = 15  # mg/L
plt.axhline(y=toxic_level, linestyle='--', label="Toxic Threshold")
plt.legend()
plt.show()

```

```python
dose2 = 200
concentration2 = (dose2 / volume) * np.exp(-elimination_rate * t)

plt.plot(t, concentration, label="100 mg dose")
plt.plot(t, concentration2, label="200 mg dose")
plt.xlabel("Time (hours)")
plt.ylabel("Drug Concentration (mg/L)")
plt.title("Effect of Dosage on Drug Concentration")
plt.legend()
plt.show()
```
